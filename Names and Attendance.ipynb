```python
import pandas as pd
```

```python
tracking = pd.read_csv('overview.csv')
```

```python
col = ['Name', 'SchoolWeek', 'FinalsWeek', 'MobilePantry', 'FoodRecovery', 'Misc', 'Total']
def entry(me):
    name = me.iloc[0]
    schoolWeek = pd.to_numeric(me.iloc[4:14], errors='coerce').sum()
    finals = int(me.iloc[14])
    mobile = int(me.iloc[16])
    food = int(me.iloc[18])
    if pd.isna(me.iloc[19]):
        M = 0
    else:
        M = int(me.iloc[19])
    tot = int(me.iloc[20])
    entryDf = pd.DataFrame([[name, schoolWeek, finals, mobile, food, M, tot]], columns = col)
    return(entryDf)
```

```python
def overview(tracking):
    test = tracking.iloc[21:,1]
    if test.isna().any():
        first_nan_index = test[test.isna()].index[0]
    df = tracking.iloc[20:first_nan_index,1:22]
    overviewdf = pd.DataFrame(columns = col)
    for i in range(len(df)):
        me = df.iloc[i,:]
        overviewdf = pd.concat([overviewdf, entry(me)])
    return(overviewdf)
```

```python
df = overview(tracking)
```

```python
col2 = ['name', 'score']
def reputation(df):
    reputationdf = pd.DataFrame(columns = col2)
    for i in range(len(df)):
        name = df.iloc[i,0]
        score = df.iloc[i,1]/2 + df.iloc[i,2] * 2 + df.iloc[i,3] * 2 + df.iloc[i,4] * 2 + df.iloc[i,5]
        entryDf = pd.DataFrame([[name, score]], columns = col2)
        reputationdf = pd.concat([reputationdf, entryDf])
    return(reputationdf)
        
```

```python
test = reputation(df)
```
